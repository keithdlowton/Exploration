```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plotting pretty figures and avoid blurry images
%config InlineBackend.figure_format = 'retina'
# Larger scale for plots in notebooks
sns.set_context('notebook')

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Enable multiple cell outputs
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# Load sample data
cars = sns.load_dataset('mpg')
stocks = pd.read_csv('prices-split-adjusted.csv',
                     parse_dates=['date'],
                     index_col=0)
```

```python
cars.head()
cars.info()
cars.describe()
```

```python
stocks.head()
stocks.info()
```

```python
cars.dropna(inplace=True)
stocks.dropna(inplace=True)
```

<h2> Scatter plots

```python
sns.relplot(x='weight', y='horsepower', 
            data=cars, kind='scatter');
```

<h3> Size of points relating to third variable

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement');
```

```python
sns.relplot(x='displacement',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='weight');
```

<h3> Size of points relating to third variable and colour of points relatig to fourth variable

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='acceleration');
```

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='origin');
```

```python
cars['round_acceleration'] = round(cars['acceleration'],0)
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='round_acceleration',
            palette='tab10');      # change of palette

# http://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
sns.relplot(x='weight', y='horsepower', 
            data=cars, hue='origin');
```

<h3> Changing style of markers

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            hue='origin',
            style='origin');
```

```python
sns.relplot(x='weight', y='horsepower', 
            data=cars, style='origin');
```

```python
hue_colors = {'usa': 'red', 
              'japan': 'orange', 
              'europe': 'green'}
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            hue='origin',
            style='origin',
            palette=hue_colors);
```

```python
sns.relplot(x='weight', y='horsepower', 
            data=cars, alpha=0.6);
```

<h2> Subplots

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='round_acceleration',
            palette='tab10',
            col='origin');
```

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='round_acceleration',
            palette='tab10',
            col='origin',
            col_wrap=2);
```

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='round_acceleration',
            palette='tab10',
            col='origin',
            col_order=['japan', 'europe', 'usa']);
```

```python
sns.relplot(x='weight',
            y='horsepower',
            data=cars,
            kind='scatter',
            size='displacement',
            hue='round_acceleration',
            palette='tab10',
            row='origin');
```

<h2> Line plots

```python
stocks['x'] = stocks.index
sns.relplot(x='x', y='close', 
            data=stocks, kind='line');
```

```python
am_ap_go = stocks[stocks['symbol'].isin(['AMZN', 'AAPL', 'GOOGL'])]

sns.relplot(x='x',
            y='close',
            data=am_ap_go,
            kind='line',
            hue='symbol');
```

```python
sns.relplot(x='x',
            y='close',
            data=am_ap_go,
            kind='line',
            hue='symbol',
            style='symbol');
```

<h3> Different markers

```python
smaller_subset = am_ap_go.loc["2015-01-01":"2015-03-01"]

sns.relplot(x='x',
            y='close',
            data=smaller_subset,
            kind='line',
            hue='symbol',
            style='symbol',
            markers=True)
plt.xticks(rotation=60);
```

<h3> Confidence intervals

```python
sns.relplot(x='x', y='close', 
            data=am_ap_go, kind='line');
```

```python
sns.relplot(x='x',  
            y='close', 
            data=am_ap_go,  
            kind='line', ci=None);
```

```python
### standard deviation instead of the confidence interval ###

sns.relplot(x='x', 
            y='close', 
            data=am_ap_go, 
            kind='line', 
            ci='sd');
```
