```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scalecast.Forecaster import Forecaster
```

```python
from scalecast.SeriesTransformer import SeriesTransformer
```

#### First Example

```python
data = pd.read_csv('D:\\PythonProjects\\DataSets\\AirPassengers.csv', parse_dates = ['Month'], index_col = ['Month'])
```

```python
data = data.sort_values(['Month'])
```

```python
data.head()
```

```python
type(data)
```

```python
f_trans = Forecaster(
    y = data.iloc[:, 0] ,   #passengers column 0th col
    current_dates = data.index, #index column
    future_dates = 5,
)
```

```python
f_trans
```

```python
f_trans.plot()
plt.show()
```

```python
transformer = SeriesTransformer(f_trans)
```

```python
f_log = transformer.LogTransform()
f_log = transformer.DetrendTransform()
```

```python
f_log.plot()
plt.show()
```

```python
# these will all be reverted later after forecasts have been called
f_trans = transformer.DiffTransform(1)
f_trans = transformer.DiffTransform(1)
f_trans = transformer.DetrendTransform()
```

```python
f_trans.plot()
plt.show()
```

#### Third Example

```python
# read in data
data = pd.read_csv('D:\\PythonProjects\\DataSets\\avocado.csv',parse_dates = ['Date'])
```

```python
data.head()
```

```python
data = data.sort_values(['region','type','Date'])
```

```python
data.head()
```

```python
volume = data.groupby('Date')['Total Volume'].sum()
```

```python
volume
```

```python
#Scalecast installation
#pip install --upgrade scalecast   
from scalecast.Forecaster import Forecaster
```

```python
f_trans = Forecaster(
    y = volume ,
    current_dates = volume.index,
    future_dates = 13,
)
```

```python
f_trans
```

```python
f_trans.plot()
plt.show()
```

```python
f_trans.set_test_length(0.15)
f_trans.set_validation_length(13)
```

```python
from scalecast.SeriesTransformer import SeriesTransformer
```

```python
transformer = SeriesTransformer(f_trans)
```

```python
# these will all be reverted later after forecasts have been called
f_trans = transformer.DiffTransform(1)
f_trans = transformer.DiffTransform(52)
f_trans = transformer.DetrendTransform()
```

```python
f_trans.plot()
plt.show()
```

```python
f_trans.add_time_trend()
f_trans.add_seasonal_regressors('week',sincos=True,raw=False)
f_trans.add_ar_terms(13)
```

```python
# f_trans.set_estimator('xgboost')
# f_trans.manual_forecast(gamma=1,dynamic_testing=13)

f_trans.set_estimator('mlr')
f_trans.manual_forecast(dynamic_testing=13)

# f.set_estimator('lasso')
# f.manual_forecast(alpha=0.2,dynamic_testing=13)

# f.set_estimator('ridge')
# f.manual_forecast(alpha=0.2,dynamic_testing=13)

# f.set_estimator('elasticnet')
# f.manual_forecast(alpha=0.2,l1_ratio=0.5,dynamic_testing=13)

# f.set_estimator('sgd')
# f.manual_forecast(alpha=0.2,l1_ratio=0.5,dynamic_testing=13)
```

```python
f_trans.plot_test_set()
plt.show()
```

```python
f_trans.plot()
plt.show()
```

```python
# call revert functions in the opposite order as how they were called when transforming
f_trans = transformer.DetrendRevert()
f_trans = transformer.DiffRevert(52)
f_trans = transformer.DiffRevert(1)
```

```python
f_trans.plot_test_set()
plt.show()
```

```python
f_trans.plot()
plt.show()
```
