```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

from statsmodels.tsa.forecasting.theta import ThetaModel
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

### Functions

Since our test set has 104 weeks of data, we will perform rolling forecasts. This means that we train the model on the training set, predict the next 52 weeks, then retrain the model on an updated training set that includes another 52 weeks of data to forecast the next 52 weeks.

From the figure above, we can see that we forecast the entire test set of 104 weeks, but 52 weeks at a time. That way, we simulate the process of collecting new data and adding to the training set before making new predictions.

```python
def rolling_forecast(df: pd.DataFrame, train_len: int, horizon: int, window: int, method: str) -> list:
    
    total_len = train_len + horizon
    end_idx = train_len
    if method == 'last_season':
        pred_last_season = []
        
        for i in range(train_len, total_len, window):
            last_season = df[:i].iloc[-window:].values
            pred_last_season.extend(last_season)
            
        return pred_last_season
    
    elif method == 'theta':
        pred_theta = []
        
        for i in range(train_len, total_len, window):
            tm = ThetaModel(endog=df[:i], period=52)
            res = tm.fit()
            predictions = res.forecast(window)
            pred_theta.extend(predictions)
            
        print(res.summary()) # Optional
        return pred_theta
    
    elif method == 'tes':
        pred_tes = []
        
        for i in range(train_len, total_len, window):
            tes = ExponentialSmoothing(
                df[:i],
                trend='add',
                seasonal='add',
                seasonal_periods=52,
                initialization_method='estimated'
            ).fit()
            
            predictions = tes.forecast(window)
            pred_tes.extend(predictions)
        return pred_tes
```

```python
def mape(y_true, y_pred):
    return round(np.mean(np.abs((y_true - y_pred) / y_true)) * 100, 2)
```

### Read in the data

```python
### Load the data ###

df = sm.datasets.co2.load_pandas().data
df.head()
```

```python
fig, ax = plt.subplots()
ax.plot(df['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 concentration (ppmw)')
fig.autofmt_xdate()
plt.tight_layout()
```

* Our series has a positive trend, as CO2 concentration is increasing over time.  
* Our series has a yearly seasonality, as CO2 is higher during the winter than in the summer.  
* We have some missing values at the beginning of the dataset.

```python
### Interpolate missing values ###

df = df.interpolate()
```

```python
### Test/train split ###

train = df[:-104]
test = df[-104:]
```

```python
### Forecast ###

TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 52

pred_tes = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'tes')
test.loc[:, 'pred_tes'] = pred_tes
test.head()
```

```python
pred_theta = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'theta')
pred_last_season = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'last_season')

test.loc[:, 'pred_last_season'] = pred_last_season
test.loc[:, 'pred_theta'] = pred_theta
test.head()
```

```python
def convert(row):
  row['pred_last_season'] = pd.Series(row['pred_last_season'])
  return row

test = test.apply(convert, axis=1)
test['pred_last_season'] = pd.Series(test['pred_last_season'])
test.head()
```

```python
fig, ax = plt.subplots()
ax.plot(df['co2'])
ax.plot(test['co2'], 'b-', label='actual')
ax.plot(test['pred_last_season'], 'r:', label='baseline')
ax.plot(test['pred_theta'], 'g-.', label='Theta')
ax.plot(test['pred_tes'], 'k--', label='TES')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 concentration (ppmv)')
ax.axvspan('2000-01-08', '2001-12-29', color='#808080', alpha=0.2)
ax.legend(loc='best')
fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot(df.loc[df.index > '1998-03-07','co2'])
ax.plot(test['co2'], 'b-', label='actual')
ax.plot(test['pred_last_season'], 'r:', label='baseline')
ax.plot(test['pred_theta'], 'g-.', label='Theta')
ax.plot(test['pred_tes'], 'k--', label='TES')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 concentration (ppmv)')
ax.axvspan('2000-01-08', '2001-12-29', color='#808080', alpha=0.2)
ax.legend(loc='best')
fig.autofmt_xdate()
plt.tight_layout()
```

```python
mape_baseline = mape(test['co2'], test['pred_last_season'])
mape_theta = mape(test['co2'], test['pred_theta'])
mape_tes = mape(test['co2'], test['pred_tes'])

fig, ax = plt.subplots()
x = ['Baseline', 'Theta', 'TES']
y = [mape_baseline, mape_theta, mape_tes]
ax.bar(x, y, width=0.4)
ax.set_xlabel('Exponential smoothing models')
ax.set_ylabel('MAPE (%)')
ax.set_ylim(0, 1)
for index, value in enumerate(y):
    plt.text(x=index, y=value + 0.05, s=str(value), ha='center')
    
plt.tight_layout()
```
